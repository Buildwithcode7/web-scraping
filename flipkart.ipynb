```python
name=[]
rating=[]
display=[]
cam=[]
battery=[]
processor=[]
prices=[]

for j in range (1,46):
    print(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    conn=requests.get('https://www.flipkart.com/search?q=smartphones&sort=price_desc&page={}'.format(j),headers=headers).text
    soup=BeautifulSoup(conn,'lxml')
    company=soup.find_all('div',class_="ZFwe0M row")
    for i in company:
            name_value=(i.find('div',class_="RG5Slk").text.strip())
            name.append(name_value)
            rating_tag = i.find('div', class_="MKiFS6")
            
            if rating_tag:
                rating.append(rating_tag.text.strip())
            else:
                rating.append(None)
            prices.append(int(i.find('div', class_="hZ3P6w DeU9vF").text.strip().replace('₹','').replace(',','')))
            display.append(i.find_all('li',class_="DTBslk")[1].text.strip())
            cam.append(i.find_all('li',class_="DTBslk")[2].text.strip())
            processor_value=(i.find_all('li',class_="DTBslk")[-2])
            if processor_value:
                processor.append(processor_value.text.strip())
            else:
                rating.append(None)   
            battery.append(i.find_all('li',class_="DTBslk")[3].text.strip())
           
```

```python
df=pd.DataFrame({
    "name":name,
    "rating":rating,
    "display":display,
    "camera":cam,
    "battery":battery,
    "battery":battery,
    "processors":processor,
    "prices":prices,
    
})    
```

```python
df.isnull().sum()
df=df.fillna(' Not available')
df=df.replace('\n','',regex=True)
```

```python
df.to_csv("output.csv", index=False)
```
